## 2. Implementing Gradient Descent Algorithm (60 pts)

Given a function $f(x) = \sqrt{x^2+5}$, $x\in \mathbf{R}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x).

def f(x):
    return (x**2+5)**0.5

# Plot the function f(x), np.linspace might be helpful.
x = np.linspace(-5, 5, 500)
plt.plot(x,f(x), label=r"y=$\sqrt{x^2+5}$", color = 'blue')
plt.title("Gradient Descent to Find Minimum")
plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.grid()
plt.legend()
plt.show()

```

```python
# Define the derivative of the function f(x), for gradient descent
def d(x):
    return x*((x**2+5)**(-0.5))

# write a function to run the Gradient Descent Algorithm, take input arguments x_0, step_size, and num_iterations; return a list/array of x_k values.
def gradient_descent(x_0, f, step_size, num_iterations):
    t = 0
    ret = []
    x = x_0
    while t < num_iterations:
        ret.append(x)
        x = x - step_size*f(x)
        t = t+1;
    return ret

# set up/assign initial values (sbu_id = 116743079)
x_0 = 9%5+1
num_iterations = 50
step_size = 1

# Run gradient descent by calling the function
g = gradient_descent(x_0, d, step_size, num_iterations)

# report x_T by printing out the value
x_T = g[-1]
print(f"x_T = {x_T: .2f}") 
# Plot the f(x_k) values vs. iteration number
h = [f(x) for x in g]

plt.plot(range(num_iterations), h, color="b")
plt.title(r"$f(x_k)$ vs. Iteration number")
plt.xlabel(r"k")
plt.ylabel(r"$f(x_k)$")
plt.grid()
plt.show()
```

```python
# create a list containing step sizes
step_sizes = [5,3,1,0.5]
# initialize a list to store the results from different step sizes
results = []

# loop through the step sizes and perform GD and record results
for step_size in step_sizes:
    results.append(gradient_descent(x_0, d, step_size, num_iterations))

results = np.array(results)

diff =[]
for i in range(len(step_sizes)):   
    diff.append(results[i][:-1] - results[i][1:])

# plot multiple curves in a single graph
plt.figure(figsize=(12, 6), dpi=80)
for i in range(len(step_sizes)):   
    plt.plot(range(num_iterations-1), diff[i], label=f"step size = {step_sizes[i]}")
plt.title(r"$f(x_{k-1}) - f(x_{k})$ vs. Iteration number")
plt.xlabel(r"k")
plt.ylabel(r"$f(x_{k-1}) - f(x_{k})$")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(79)   # change the seed to be the last two digits of your SB ID (sbu_id = 116743079)
X = 2 * np.random.rand(100, 2)  # Generate 100 random 2D data points
y = 4 + 4 * X[:, 0] + 0.5 * X[:, 1] + np.random.randn(100)  # Generate y values with noise

# Add a column of ones to X for the bias terms (intercept) in the normal equation
X = np.concatenate([np.ones((len(X), 1)), X], axis=1)

X_T = np.transpose(X)
X_T_X = np.dot(X_T,X)
X_T_y = np.dot(X_T,y)

theta = np.linalg.solve(X_T_X, X_T_y)

intercept, *coefs = theta
print(f"Intercept: {intercept}\n\
Coefficients: {coefs}")

# Plot the generated data and the regression plane
# Create a grid for the regression plane
x1_range = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x2_range = np.linspace(X[:, 2].min(), X[:, 2].max(), 20)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

# Compute corresponding y values on the grid
y_pred = intercept + coefs[0] * x1_grid + coefs[1] * x2_grid

# Plot the data and the regression plane
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data
ax.scatter(X[:, 1], X[:, 2], y, color='blue', label='Data Points')

# Plot the regression plane
ax.plot_surface(x1_grid, x2_grid, y_pred, alpha=0.5, label='Regression Plane')

# Add labels and title
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Linear Regression Plane')
plt.legend()
plt.show()
```

## 4. Logistic Regression with Numpy and Graphing with Matplotlib (20 pts)

In class, we implemented a logistic regression model to perform binary classification. Here, I provided the exact same code that we discussed in class, but the values for learning_rate and num_epochs are modified. I also removed all the comments. You task is to
1. Tune the Learning Rate and Number of Epochs: Your first task is to fine-tune the values for the learning_rate and num_epochs parameters. Your goal is to identify suitable values that will enable us to converge to a set of parameter values $\theta$ closely approximating the optimal parameter values $\theta^*$.
2. Code Explanation: To enhance code comprehension, please augment the code with meaningful comments. These comments should elucidate the purpose and functionality of each code segment, making it easier for readers to understand the logistic regression implementation.

By accomplishing these tasks, we aim to achieve a better understanding of the logistic regression model's behavior and its parameter optimization process.

```python
import numpy as np
import matplotlib.pyplot as plt

# sigmoid activation function. 
#   maps inputs to a range between 0 and 1
def sigmoid(z):
    return 1 / (1 + np.exp(-z))


# Generating  Data
np.random.seed(79) # Set a random seed for reproducibility
X = np.random.rand(2, 100) # Generate 100 2D random samples in [0, 1]
w_true = np.array([1.5, -2.5]) # True weights for generating labels
b_true = 1.0  # True bias
probabilities = sigmoid(np.dot(w_true.T, X) + b_true) # Compute true probabilities
Y = (probabilities > 0.5).astype(int) # Generate binary labels (0 or 1) based on probabilities

# Adding Noise to Features
# make the data more realistic and challenging for the model.
X = X + 0.3 * np.random.rand(2, 100) - 0.1 * np.random.rand(2, 100)

# Splitting Data into Training and Test Sets
X_train, X_test = X[:,:80], X[:,80:]
Y_train, Y_test = Y[:80], Y[80:]

# Initialize weights and bias
w = np.zeros(X.shape[0])
b = 0.0

# Fine-tuned learning rate and number of epochs
learning_rate = 7.9
num_epochs =12

# Training the Logistic Regression Model
for epoch in range(num_epochs):
    # Compute the predicted probabilities for the training set
    A_train = sigmoid(np.dot(w.T, X_train) + b)

    # Gradients of the cost function (log loss) with respect to weights (dJdw) and bias (dJdb) are calculated.
    dJdw = np.dot(X_train, (A_train - Y_train).T) / len(Y_train)
    dJdb = np.mean(A_train - Y_train)

    # Parameters (w and b) are updated using gradient descent
    w -= learning_rate * dJdw
    b -= learning_rate * dJdb


# Computing Predictions and Accuracy

# Predict probabilities for the training and test set and convert to binary predictions
A_train = sigmoid(np.dot(w.T, X_train) + b)
predictions_train = (A_train > 0.5).astype(int)

A_test = sigmoid(np.dot(w.T, X_test) + b)
predictions_test = (A_test > 0.5).astype(int)

# Calculate training and test accuracies
train_accuracy = np.mean(predictions_train == Y_train)
test_accuracy = np.mean(predictions_test == Y_test)

print(f"Training Set Accuracy: {train_accuracy:.2f}")
print(f"Test Set Accuracy: {test_accuracy:.2f}")
```

```python
# Extract the first 80 samples (training data) and corresponding labels for the two features.
plt.scatter(X[0, :80], X[1,:80], c=Y[:80], cmap=plt.cm.Paired)

plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)
ax = plt.gca()

xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid of points covering the specified x and y ranges.
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))

# Flattens the grid points into pairs for computation
# and applies the logistic regression model to compute the linear combination of weights, inputs, and bias
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b

Z = sigmoid(Z) # convert the outputs into probabilities.
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

# Plot the Decision Boundary
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```

```python
# Extract the last 20 samples (testing data) and corresponding labels for the two features.
plt.scatter(X[0, 80:], X[1, 80:], c=Y[80:], cmap=plt.cm.Paired)
plt.xlim(-0.2, 1.2)
plt.ylim(-0.2, 1.2)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create a grid of points covering the specified x and y ranges.
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))

# Flattens the grid points into pairs for computation
# and applies the logistic regression model to compute the linear combination of weights, inputs, and bias
Z = np.dot(w.T, np.c_[xx.ravel(), yy.ravel()].T) + b
Z = sigmoid(Z) # convert the outputs into probabilities.
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[0.5], alpha=0.5, linestyles=['--'])

# Plot the Decision Boundary
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')
plt.show()
```
